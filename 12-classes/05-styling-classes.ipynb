# Classes

## Styling Classes

While different projects adopt different styles, here are a few tips that work as a nice default.
- Classes should be named in *CamelCase*. To do this, capitalize the first letter of each word in the name, and don't use underscores.
- **Always** include a docstring following the class definition that briefly explains what it does.
- Use blank lines to organize your code, but don't use them excessively. A good rule of thumb is one blank space between class methods, and two blank lines between classes in the same file.
- When importing modules from the Python standard library and modules that you've written, include the Python standard library imports first. Then add a blank line, and import the module you wrote.
